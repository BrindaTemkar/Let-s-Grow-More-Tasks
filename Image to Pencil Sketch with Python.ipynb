```python
# Step 1: Importing the opencv and matplotlip libraries 
import cv2
import matplotlib.pyplot as plt
```

```python
#command used to read image using OpenCv
imge = cv2.imread('GongYoo.jpg')
```

```python
# This opens the image in a new window named 'original_image'
cv2.imshow('original_image',imge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# If we want to diplay the image on the Jupyter Notebook we have to use matplotlib'spyplot
plt.imshow(imge)
plt.axis(False)
plt.show()
```

The color of the image opened using matplotlib is different  from the open opened using OpenCv because matpotlib uses the RGB color scheme whereas OpenCv uses BGR color scheme

```python
# To convert the image from BGR to RGB format
#plt.imshow(imge[:,:,::-1])
#plt.axis(False)
#plt.show()
RGB_img = cv2.cvtColor(imge, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

Now that we have loaded the image in Jupyter Notebook we proceed with the next steps.

```python
# Step 2: Convert the given colored image into a black-and-white image.
#grey_img = cv2.imread(imge, cv2.IMREAD_GRAYSCALE)
#print(img.shape)
#plt.imshow(grey_img, cmap="gray")
grey_img = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)

# Step 3 : Invert image
invert_imge = cv2.bitwise_not(grey_imge)

# Step 4 : Blur the image
blur_imge = cv2.GaussianBlur(invert_imge, (111,111), 0)

# invert the blurred image
blur_invert= cv2.bitwise_not(blur_imge)
```

```python
# Step 5 : Sketch
# The Sketch can be obtaine be performing bitwise divide between the grey image and the inverted blur image.
sketch = cv2.divide(grey_img, blur_invert, scale = 256.0)
sketch_RGB = cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB)
```

```python
## Original vs Sketch 
plt.figure(figsize = (14,8))

plt.subplot(1,2,1)
plt.title('Original Image',size = 18)
plt.imshow(RGB_img)

plt.subplot(1,2,2)
plt.title('Sketch',size = 18)
plt.imshow(sketch_RGB)
plt.show()
```
